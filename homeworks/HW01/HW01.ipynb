***Импорт NumPy***

```python
import numpy as np
```

```python
a = np.array([1, 2, 3]) #преобразовали обычный список в np массив
print(type(a))
b = np.zeros(6) #создали массив с 3 нуоями
c = np.ones((3,4))  #создали матрицу с 3 строками и 4 столбцами, заполненную единицами
d = np.arange(20, 31, 2) # Создали массив от 20 до 31 (не включительно) с шагом 2
print(b)
print(c)
print(d)
print(c.shape) #размерность матрицы
print(c.dtype) #тип данных матрицы
print(d[0])
print(d[1:3])
print(d[::2]) #На всю область массива через один элемент
print(b + d) # сумма 0 и элемента из d равна элементу d
print(b * d) #Умножая каждый элемент массива d на 0 мы получаем массив 0
print(d / b) #Деление на 0 приводит не к ошибке, а к бесконечности (и предупреждению)
print(d.mean()) #Среднее значение массив d
```
